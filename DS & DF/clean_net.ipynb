```python
import numpy as np
import pandas as pd
import re
```

```python
# load dataset
df = pd.read_csv('netflix1.csv')
df.head()
```

```python
df.info()
```

```python
# # drop show_id 
df.drop(columns ='show_id' ,inplace = True)
```

```python
# check if columns has special character
special_charcter = df['title'].str.contains('[^a-zA-Z0-9\s]', regex=True)
special_charcter.unique().sum()
df[special_charcter]
```

```python
special_charcters = df['listed_in'].str.contains('[^a-zA-Z0-9\s]', regex=True)
special_charcters.unique().sum()
df[special_charcters]
```

```python
# remove special character
def remove_special_characters(text):
    return re.sub(r'[^a-zA-Z0-9\s]', '', text)
```

```python
# remove special character in title
df['title']  = df['title'].apply(remove_special_characters)
```

```python
df['listed_in']  = df['listed_in'].apply(remove_special_characters)
```

```python
special_charcters = df['listed_in'].str.contains('[^a-zA-Z0-9\s]', regex=True)
special_charcters.unique().sum()
df[special_charcters]
```
